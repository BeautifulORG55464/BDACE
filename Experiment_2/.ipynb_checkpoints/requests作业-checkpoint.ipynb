```python
# 代码3-6
import requests
url = 'http://www.tipdm.com'
header = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/133.0.0.0 Safari/537.36 Edg/133.0.0.0'}
# 生成GET请求
rqg = requests.get(url, headers=header)
 # 查看结果类型
print('结果类型:',type(rqg))
 # 查看状态码
print('状态码:',rqg.status_code)
 # 查看编码
print('编码类型:',rqg.encoding)
 # 查看响应头
print('响应头:',rqg.headers)
 # 查看网页内容
print('纯文本：',rqg.text)
print(rqg.content)
```

```python
# 代码3-8
import chardet
url = 'http://www.tipdm.com'
rqg = requests.get(url,headers=header)
print('编码：', rqg.encoding)  # 查看编码
print('detect()方法检测结果：', chardet.detect(rqg.content))
# 将检测到的编码赋值给rqg.encoding
rqg.encoding = chardet.detect(rqg.content)['encoding']
# 查看改变后的编码
rqg.encoding

```
