```python
# 代码3-18
from bs4 import BeautifulSoup
import requests
import chardet
url = 'http://www.tipdm.com'
ua = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/134.0.0.0 Safari/537.36 Edg/134.0.0.0'}
rqg = requests.get(url, headers=ua)
rqg.encoding = chardet.detect(rqg.content)['encoding']
# 初始化HTML对象
html = rqg.text
soup = BeautifulSoup(html, 'lxml')  # 生成BeautifulSoup对象
print('输出格式化的BeautifulSoup对象：', soup.prettify())  # 输出格式化的Beau soup.select('a[class="logo"]'))

# 使用[]获取标签的属性值
print('获取标签中的属性值：', soup.select('a[class="logo"]>img')[0]['src'])

```

```python
# 代码3-19
# 获取<head>标签
head = soup.head
print(head)
# 获取<title>标签
title = soup.title
print(title)
# 获取<body>标签中的第一个<a>标签
ba = soup.body.find('a')
print('\n`<body>`标签中的第一个`<a>`标签:',ba)
# 获取所有名称为a的标签的个数
baa = soup.body.find_all('a')
print('\n获取所有名称为a的标签的个数:',len(baa))
```

```python
# 代码3-25，补全每行的代码
# 通过name参数搜索名为title的全部子节点
print('名为title的全部子节点：', soup.find_all(name='title'))
print('title子节点的文本内容：', soup.title.string)
print('使用get_text()获取的文本内容：', soup.title.get_text()                    )

target = soup.find(class_='next',name='a')            # 查找的class类属性值为next，标签名为“a”的节点
print('获取的节点：', target)

target = soup.find_all( attrs = {'id':'search'})  # 传入关键字id，搜索id='search'的节点
print('关键字id匹配的节点：', target)

target =  soup.ul.find('a')
print('第一个ul节点下名称为a的节点：', target)

# 创建两个空列表用于存放链接及文本
urls = []
text = []
# 分别提取链接和文本
for tag in target:
    urls.append(  tag.href      )
    text.append(   tag.get_text()             )
print('a节点的href属性值构成的列表：', urls)
print('a节点的文本内容构成的列表：', text)
```

```python
# 代码3-26,select用法，补齐句子空白处代码
html = rqg.text
soup = BeautifulSoup(html, 'lxml')  # 生成BeautifulSoup对象
# 通过标签名查找
print('查找标签名为title的标签：',     soup.select('title')                           )
# 使用text获取列表中标签元素的文字
print('获取标签中文字：',       soup.select('title')[0].get_text()                                  )
# 使用类名查找
print('查找类名为slogan的标签：',            soup.select('.slogan')                           )

# 使用id名查找 
print('查找id名为menu的标签：',      soup.select('#menu')                                   )
 
# 使用组合查找
print('查找<div>标签的类名为introTxt的标签：',       soup.select('div#introTxt')                    )

# 使用子标签查找
print('查找父标签为head的<title>标签：',       soup.head.select('title',recursive=False)                      )

# 使用属性查找
print('查找属性class="logo"的<a>标签：',    soup.select('a.logo')                         )

# 使用[]获取标签的属性值
print('获取选择class属性为logo的<a>标签的子标签中的src属性的值：',     soup.select('a.logo')[0].href                      )
```
